
# Week 8 - Single Agent System & Agent Pipeline

## Objective

The objective of this project is to build a simple single-agent pipeline capable of routing user queries to different tools based on predefined rules.

The agent analyzes the user's input, identifies the query type, selects the appropriate tool, and returns the corresponding response.
---



````python
# Calculator Tool
def calculator(expression):
    try:
        return eval(expression)
    except:
        return "Invalid Expression"


# Keyword Extraction Tool
def keyword_extractor(text):
    words = text.split()
    keywords = [word for word in words if len(word) > 4]
    return keywords


# General Response Tool
def general_response(query):
    return "This is a general response from the agent."
````

````python
def agent_router(query):

    query_lower = query.lower()

    if "calculate" in query_lower:
        expression = query_lower.replace("calculate", "").strip()
        return calculator(expression)

    elif "keywords" in query_lower:
        text = query.replace("keywords", "")
        return keyword_extractor(text)

    else:
        return general_response(query)
````

````python
queries = [
    "Calculate 45+40",
    "Extract keywords from Artificial Intelligence is transforming education",
    "Tell me about Python"
]

for q in queries:
    print("Query :", q)
    print("Response :", agent_router(q))
    print("-"*60)
````

# Results

The single-agent system successfully routed user queries based on predefined rules.

The implemented agent was able to:

- Perform arithmetic calculations.
- Extract keywords from text.
- Respond to general queries.

The routing mechanism correctly selected the appropriate tool for each query, demonstrating the working of a simple rule-based agent pipeline.
